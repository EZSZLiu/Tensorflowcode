```python
import pandas as pd
import numpy as np
df=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data')
#提取特征标签为numpy数组形式
y=df.iloc[0:100,4].values
#np.where(condition,x,y)满足输出x，否则y
y=np.where(y=='Iris-setosa',1,-1)
#提取x两个feature,转array格式。不加values为DF
x=df.iloc[:100,[0,2]].values
x
```

```python
#感知机模型
from sklearn.linear_model import Perceptron 
ppn=Perceptron()
ppn.fit(x,y)
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
def data_visiualization(x,y,classifier,resolution):
    makers=('s','x','o','^','y')
    colors=('red','blue','lightgreen','gray','cyan')
    cmap=ListedColormap(colors[:len(np.unique(y))])
    #plot the dicision surface
    x_1min,x_1max=x[:,0].min()-1,x[:,0].max()+1
    x_2min,x_2max=x[:,1].min()-1,x[:,1].max()+1
    #生成网格
    xx1,xx2=np.meshgrid(np.arange(x_1min,x_1max,resolution),np.arange(x_2min,x_2max,resolution))
    z=classifier.predict(np.array([xx1.ravel(),xx2.ravel()]).T)  #默认行向量
    

    z=z.reshape(xx1.shape)
    plt.contourf(xx1,xx2,z,alpha=0.3,cmap=cmap)
    #坐标上下限
    plt.xlim(xx1.min(),xx1.max())
    plt.ylim(xx2.min(),xx2.max())
#     i=0
    for idx,cl in enumerate(np.unique(y)):
#         print('cl',cl)
        plt.scatter(x=x[y==cl,0],y=x[y==cl,1],alpha=0.8,c=colors[idx],label=cl,edgecolor='black')
#         print('第%d次x'%i,x)
#         print('第%d次y'%i,y)
#         i+=1
```

```python
data_visiualization(x,y,classifier=ppn,resolution=0.02)
plt.xlabel('**')
plt.ylabel('__')
plt.legend(loc='upper_left')
plt.show()
```
